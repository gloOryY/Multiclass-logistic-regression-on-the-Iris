```python
X = df.drop('target', axis=1)  
# X – матрица признаков (features)
# df.drop('target', axis=1) – удаляем столбец 'target' по оси столбцов (axis=1)
# В X остаются только входные признаки: age, sex, cp, trestbps, chol, ...
```

```python
y = df['target']  
# y – целевая переменная (labels)
# Это столбец, который мы хотим предсказывать: 0 – нет болезни, 1 – есть болезнь
```

```python
from sklearn.model_selection import train_test_split  
# Импорт функции, которая делит данные на обучающую и тестовую части
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y,            # все признаки и все метки
    test_size=0.25,  # 25% пойдёт в тест, 75% — в обучение
    random_state=101 # "сид" генератора случайных чисел — для воспроизводимости
)
# X_train, y_train – обучающая выборка, на ней учим модель
# X_test, y_test – тестовая выборка, на ней проверяем качество на "новых" данных
```

```python
from sklearn.preprocessing import StandardScaler  
# Импорт стандартизатора признаков (масштабирование к нулевому среднему и единичному отклонению)
```

```python
scaler = StandardScaler()  
# Создаём объект масштабировщика, пока без данных
```

```python
scaled_X_train = scaler.fit_transform(X_train)
# fit_transform:
# 1) fit: считает mean и std по каждому признаку на обучающей выборке
# 2) transform: применяет формулу (x - mean) / std → стандартизированные признаки
# Результат: scaled_X_train – numpy-массив того же размера, но в стандартизированном масштабе
```

```python
scaled_X_test = scaler.transform(X_test)
# Тут только transform, без fit!
# Мы используем mean и std, посчитанные по X_train
# Нельзя "подглядывать" в тест → иначе утечка информации 
```

```python
from sklearn.linear_model import LogisticRegression  
# Импорт класса логистической регрессии — линейная модель для задач классификации.
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
# Импортируем метрики качества:
# - accuracy_score      → доля правильных ответов
# - confusion_matrix    → матрица ошибок (TP, FP, FN, TN)
# - classification_report → подробный отчёт по каждому классу (precision, recall, f1)
```

```python
log_model = LogisticRegression()
# Создаём объект модели с параметрами по умолчанию
# Важные параметры (по умолчанию):
# - penalty='l2'        → L2-регуляризация (борьба с переобучением)
# - C=1.0               → сила регуляризации (обратная, больше C → слабее штраф)
# - solver='lbfgs'      → алгоритм оптимизации (подбирает веса модели)
# - max_iter=100        → максимальное число итераций для сходимости

log_model.fit(scaled_X_train, y_train) # обучение на train

y_pred = log_model.predict(scaled_X_test)
# Делаем предсказания на тестовой выборке
# scaled_X_test — новые (невиданные при обучении) примеры
# y_pred — массив предсказанных классов (0/1)
```

```python
accuracy_score(y_test, y_pred)
# Считаем accuracy: (количество верных предсказаний) / (общее число объектов)
```

```python
confusion_matrix(y_test, y_pred)
# Строим матрицу ошибок:
# [[TN, FP],
#  [FN, TP]]
# где:
#  TN — верно предсказали 0
#  TP — верно предсказали 1
#  FP — ошибочно предсказали 1 вместо 0
#  FN — ошибочно предсказали 0 вместо 1
```

```python
print(classification_report(y_test, y_pred))
# Печатаем отчёт по каждому классу (0 и 1):
# - precision: из всех предсказанных "1" сколько действительно 1
# - recall: из всех реальных "1" сколько мы нашли
# - f1-score: баланс precision и recall
# - support: сколько объектов этого класса в тесте
```

```python
# Оцениваем качество
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Confusion matrix:\n", confusion_matrix(y_test, y_pred))
print("Report:\n", classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import roc_curve, roc_auc_score
# roc_curve  — точки для ROC-кривой (FPR, TPR)
# roc_auc_score — площадь под ROC-кривой (AUC)
```

```python
# Получаем вероятности принадлежности к классу 1
y_prob = log_model.predict_proba(scaled_X_test)[:, 1]
# predict_proba(X) - массив размера (n_samples, 2):
#   [:, 0] — p(class=0)
#   [:, 1] — p(class=1)
# Берём второй столбец: вероятность того, что объект относится к классу 1
```

```python
fpr, tpr, thresholds = roc_curve(y_test, y_prob)
# roc_curve:
# - y_test    — истинные метки (0/1)
# - y_prob    — предсказанные вероятности класса 1
# Возвращает:
# - fpr: False Positive Rate (доля объектов класса 0, ошибочно принятых за 1)
# - tpr: True Positive Rate (recall: доля объектов класса 1, правильно найденных)
# - thresholds: пороги, при которых считались fpr и tpr
```

```python
roc_auc = roc_auc_score(y_test, y_prob)
# Считаем AUC (Area Under Curve) — площадь под ROC-кривой
# Значения:
#   0.5  — модель как случайный угадыватель
#   1.0  — идеальный классификатор
#   обычно >0.7, >0.8 — уже неплохо
```

```python
plt.figure(figsize=(6, 6))
plt.plot(fpr, tpr, label=f'ROC curve (AUC = {roc_auc:.2f})')
# Рисуем ROC-кривую: по X — FPR, по Y — TPR

plt.plot([0, 1], [0, 1], 'k--', label='Random')
# Диагональ "случайного" классификатора (линия безразличия)

plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC curve')
plt.legend(loc='lower right')
plt.show()
# График показывает, как меняется качество модели при разных порогах классификации
```
